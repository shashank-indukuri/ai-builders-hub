### üéØ Types of Machine Learning:
1. **üéì Supervised Learning** ‚Üê *We'll build this today!*
   - Has "answer key" (labeled data)
   - Examples: Email spam detection, price prediction

2. **üîç Unsupervised Learning**
   - No "answer key" - finds hidden patterns
   - Examples: Customer segmentation, anomaly detection

3. **üéÆ Reinforcement Learning**
   - Learns through trial and error
   - Examples: Game AI, autonomous vehicles

## üìä Creating Our Dataset

### üé¨ Let's Create a Movie Dataset Together!

````python
# Let's create our movie dataset!
print("Help me build our dataset!")
print("\nShout out your favorite movies and I'll add them!")


movies_data = {
    'movie': ['Avengers: Endgame', 'Titanic', 'The Godfather', 'Toy Story', 'Inception',
              'The Dark Knight', 'Forrest Gump', 'Pulp Fiction', 'The Lion King', 'Avatar'],
    'genre': ['Action', 'Romance', 'Drama', 'Animation', 'Sci-Fi',
              'Action', 'Drama', 'Crime', 'Animation', 'Sci-Fi'],
    'duration_minutes': [181, 195, 175, 81, 148, 152, 142, 154, 88, 162],
    'budget_millions': [356, 200, 6, 30, 160, 185, 55, 8, 45, 237],
    'rating': [8.4, 7.8, 9.2, 8.3, 8.8, 9.0, 8.8, 8.9, 8.5, 7.8]
}

# Create DataFrame
df = pd.DataFrame(movies_data)

print(f"\nüìä Dataset created with {len(df)} movies!")
print("\nüéØ Our Mission: Predict movie ratings based on genre, duration, and budget!")

# Display the data
df
````

### üîç Let's Explore Our Data!

````python
# Quick data exploration
print("üìà DATA EXPLORATION")
print("=" * 50)

print(f"üìä Dataset Shape: {df.shape[0]} movies, {df.shape[1]} features")
print(f"üé¨ Genres: {df['genre'].unique()}")
print(f"‚è±Ô∏è  Duration Range: {df['duration_minutes'].min()}-{df['duration_minutes'].max()} minutes")
print(f"üí∞ Budget Range: ${df['budget_millions'].min()}-${df['budget_millions'].max()} million")
print(f"‚≠ê Rating Range: {df['rating'].min()}-{df['rating'].max()}")

# Basic statistics
print("\nüìä Quick Stats:")
df.describe()
````

### üìä Visualizing Our Data

````python
# Create engaging visualizations
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('üé¨ Movie Dataset Exploration', fontsize=16, fontweight='bold')

# 1. Rating distribution
axes[0,0].hist(df['rating'], bins=8, color='skyblue', alpha=0.7, edgecolor='black')
axes[0,0].set_title('Rating Distribution')
axes[0,0].set_xlabel('Rating')
axes[0,0].set_ylabel('Count')

# 2. Genre vs Rating
df.boxplot(column='rating', by='genre', ax=axes[0,1])
axes[0,1].set_title('Rating by Genre')
axes[0,1].set_xlabel('Genre')

# 3. Duration vs Rating
axes[1,0].scatter(df['duration_minutes'], df['rating'], color='coral', s=100, alpha=0.7)
axes[1,0].set_title('Duration vs Rating')
axes[1,0].set_xlabel('Duration (minutes)')
axes[1,0].set_ylabel('Rating')

# 4. Budget vs Rating
axes[1,1].scatter(df['budget_millions'], df['rating'], color='lightgreen', s=100, alpha=0.7)
axes[1,1].set_title('Budget vs Rating')
axes[1,1].set_xlabel('Budget (millions)')
axes[1,1].set_ylabel('Rating')

plt.tight_layout()
plt.show()

print("\nü§î DISCUSSION: What patterns do you see? Which factors might predict higher ratings?")
````

## ü§ñ Building Our AI Model

### Step 1: Prepare the Data

````python
print("üîß STEP 1: DATA PREPARATION")
print("=" * 40)

# Convert genres to numbers (computers love numbers!)
label_encoder = LabelEncoder()
df['genre_encoded'] = label_encoder.fit_transform(df['genre'])

print("üìù Genre Encoding:")
for i, genre in enumerate(label_encoder.classes_):
    print(f"   {genre} ‚Üí {i}")

# Define features (X) and target (y)
X = df[['genre_encoded', 'duration_minutes', 'budget_millions']]
y = df['rating']

print(f"\nüéØ Features (X): {list(X.columns)}")
print(f"üéØ Target (y): {y.name}")
print(f"\nüìä Feature Matrix Shape: {X.shape}")
print(f"üìä Target Vector Shape: {y.shape}")

# Show the prepared data
print("\n‚úÖ Prepared Data Sample:")
pd.concat([X.head(), y.head()], axis=1)
````

### Step 2: Split Data (Train vs Test)

````python
print("‚úÇÔ∏è  STEP 2: TRAIN-TEST SPLIT")
print("=" * 40)

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

print(f"üìö Training Set: {len(X_train)} movies")
print(f"üß™ Test Set: {len(X_test)} movies")
print(f"\nüìä Split Ratio: {len(X_train)}/{len(X_test)} = {len(X_train)/len(X_test):.1f}:1")

print("\nü§î WHY DO WE SPLIT?")
print("   üìö Training: Teach the model")
print("   üß™ Testing: Check if it learned correctly")
print("   üö´ Never let the model see test data during training!")
````

### Step 3: Train the Model

````python
print("üéì STEP 3: TRAINING THE MODEL")
print("=" * 40)

# Create and train the model
model = LinearRegression()

print("ü§ñ Creating Linear Regression model...")
print("üìö Training on movie data...")

# This is where the magic happens!
model.fit(X_train, y_train)

print("\n‚úÖ MODEL TRAINED SUCCESSFULLY!")
print("\nüß† What the model learned:")
feature_names = ['Genre', 'Duration', 'Budget']
for feature, coef in zip(feature_names, model.coef_):
    print(f"   {feature}: {coef:.3f}")

print("\nü§î INTERPRETATION:")
print("   Positive coefficient = increases rating")
print("   Negative coefficient = decreases rating")
````

### Step 4: Test the Model

````python
print("üß™ STEP 4: TESTING THE MODEL")
print("=" * 40)

# Make predictions
y_pred = model.predict(X_test)

# Calculate accuracy metrics
mse = mean_squared_error(y_test, y_pred) # MSE: Average of squared differences - penalizes large errors heavily
mae = mean_absolute_error(y_test, y_pred) # MAE: Average of absolute differences - treats all errors equally

print(f"üìä Model Performance:")
print(f"   Mean Squared Error: {mse:.3f}")
print(f"   Mean Absolute Error: {mae:.3f}")

# Show predictions vs actual
results_df = pd.DataFrame({
    'Movie': df.loc[X_test.index, 'movie'].values,
    'Actual Rating': y_test.values,
    'Predicted Rating': y_pred,
    'Difference': abs(y_test.values - y_pred)
})

print("\nüéØ PREDICTIONS vs REALITY:")
print(results_df.round(2))

# Visualize predictions
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.7, s=100)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual Rating')
plt.ylabel('Predicted Rating')
plt.title('Predictions vs Actual Ratings')
plt.grid(True, alpha=0.3)
plt.show()
````

### Step 5: Deploy - Make Real Predictions!

````python
print("üöÄ STEP 5: DEPLOYMENT - LIVE PREDICTIONS!")
print("=" * 50)

# Save the trained model
import joblib
joblib.dump(model, 'movie_rating_model.pkl')
joblib.dump(label_encoder, 'genre_encoder.pkl')
print("üíæ Model saved successfully!")

# Load the saved model (simulating deployment)
loaded_model = joblib.load('movie_rating_model.pkl')
loaded_encoder = joblib.load('genre_encoder.pkl')
print("üìÇ Model loaded for deployment!")

def predict_movie_rating(genre, duration, budget):
    """Our deployed AI model!"""
    # Encode genre
    try:
        genre_encoded = loaded_encoder.transform([genre])[0]
    except:
        print(f"‚ùå Unknown genre: {genre}")
        return None
    
    # Make prediction
    features = [[genre_encoded, duration, budget]]
    prediction = loaded_model.predict(features)[0]
    
    return round(prediction, 1)

# Interactive predictions
print("üé¨ AUDIENCE PARTICIPATION: Let's predict some movie ratings!")
print(f"Available genres: {list(loaded_encoder.classes_)}")

# Example predictions
test_movies = [
    ("Action", 150, 200),
    ("Romance", 120, 50),
    ("Animation", 90, 100)
]

print("\nüîÆ SAMPLE PREDICTIONS:")
for genre, duration, budget in test_movies:
    rating = predict_movie_rating(genre, duration, budget)
    print(f"   üé¨ {genre} movie, {duration}min, ${budget}M ‚Üí Rating: {rating}‚≠ê")

print("\nüéâ CONGRATULATIONS! You've built and deployed an AI model!")
````
